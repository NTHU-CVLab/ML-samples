```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('bike/train.csv')
df_test = pd.read_csv('bike/test.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("whitegrid")
grid = sns.FacetGrid(df, col = 'season', row = 'holiday')
grid.map(plt.scatter, 'temp', 'count', alpha = 0.5)
grid.add_legend();
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import model_selection, preprocessing
```

```python
df.datetime = df.datetime.apply(pd.to_datetime)
df['month'] = df.datetime.apply(lambda x: x.month)
df['hour'] = df.datetime.apply(lambda x: x.hour)
df['day'] = df.datetime.apply(lambda x: x.day)
df.drop(['datetime'], 1, inplace = True)
df.head()
```

```python
X = np.array(df.drop(df[['casual', 'registered', 'count']], 1))
y = np.array(df['count'])
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.2)
```

```python
def rmsle(y, y_):
    log1 = np.nan_to_num(np.array([np.log(v + 1) for v in y]))
    log2 = np.nan_to_num(np.array([np.log(v + 1) for v in y_]))
    calc = (log1 - log2) ** 2
    return np.sqrt(np.mean(calc))
```

```python
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)
print("The RMSLE is: ", rmsle(y_test, tree.predict(X_test)))
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)
draw_X = X_test[:,5]

import matplotlib.patches as mpatches
pred_patch = mpatches.Patch(color='red', label='predict')
gt_patch = mpatches.Patch(color='blue', label='gt')

axs[0].scatter(x = draw_X, y = y_test, alpha= .5, color='blue')
axs[0].set_xlabel("temperature")
axs[0].set_ylabel("count")
axs[0].legend(handles=[gt_patch])

axs[1].scatter(x = draw_X, y = y_test, alpha= .5, color='blue')
axs[1].scatter(x = draw_X, y = tree.predict(X_test), alpha= .5, color='red')
axs[1].set_xlabel("temperature")
axs[1].legend(handles=[pred_patch, gt_patch])

plt.show()
```
