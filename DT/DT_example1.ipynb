```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
```

```python
# 將資料從 dataset 中讀取出來
iris = datasets.load_iris() 
print(iris['feature_names'])
feature_names = iris['feature_names']
```

```python
x = pd.DataFrame(iris['data'], columns=feature_names)
y = pd.DataFrame(iris['target'], columns=['target'])
iris_data = pd.concat([x,y], axis=1)
```

```python
all_name = list(feature_names)
all_name.append('target')
iris_data = iris_data[all_name]
iris_data = iris_data[iris_data['target'].isin([0,1,2])]
```

```python
# 將全部樣本分群，80% 作 training data，剩下 20% 作 test data
test_size = 0.2
X_train, X_test, y_train, y_test = \
train_test_split(iris_data[feature_names], iris_data[['target']], test_size=test_size, random_state=0)
```

```python
# 建立 decision tree, 這邊採用 information gain 的方法
tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
# 執行 training
tree.fit(X_train, y_train)
```

```python
# 執行 testing, 同時顯示出 testing data 的 accuracy
accuracy = tree.score(X_test, y_test['target'])
print('accuracy = %f' % (accuracy))
pred_y = tree.predict(X_test)
X_test['gt'] = y_test['target']
X_test['predict'] = pred_y
X_test.head(10)
```

```python
from sklearn.tree import export_graphviz
from subprocess import check_call
export_graphviz(tree, out_file='iris_tree.dot', feature_names=feature_names, filled=True)
check_call(['dot','-Tpng','iris_tree.dot','-o','iris_tree.png'])
```
